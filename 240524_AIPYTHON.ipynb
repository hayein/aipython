```python
# 데이터 파일 읽어오기

import pandas as pd
iris_data = pd.read_csv('iris.csv')
iris_data.head()
```

```python
# 데이터의 기본 정보 출력
iris_data.info()
```

```python
# 데이터의 기초 통계량 출력

iris_data.describe()
```

```python
# 품종별 개수 구하기

species_counts = iris_data['species'].value_counts()
species_counts
```

```python
# 결측치 확인하기

iris_data.isna().sum()
```

```python
# 중복 데이터 확인하기

iris_data[iris_data.duplicated]
```

```python
# 중복 데이터 모두 확인하기

```

```python
# 중복 데이터 삭제하기

df2 = iris_data.drop_duplicates()
```

```python
# 품종 열을 기준으로 합계 구하기

species_sum = df2.groupby('species').sum()
species_sum
```

```python
# 품종 열을 기준으로 평균 구하기

species_mean = df2.groupby('species').mean()
species_mean
```

```python
# 막대 그래프 그리기

import matplotlib.pyplot as plt
import seaborn as sns

????????
```

```python
# 붖꽃 종류별 꽃받침 길이의 평균에 대한 시각화

sns.barplot(x='species', y='sepal_length', data=iris_data)
plt.xlabel('Species')
plt.ylabel('Mean sepal Length')
plt.title('Mean Sepal Length by Species')
plt.show()
```

```python
# 히스토그램 그래프 그리기

iris_data.hist()
plt.show()
```

```python
# 산점도 그래프 그리기

sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris_data)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Scatter')
plt.show()
```
