```python
# 다음은 한 학생의 일주일 간의 공부 시간과 성적 데이터입니다.

# 1. 왼쪽 데이터를 이용하여 데이터프레임을 생성하세요.

df2 = pd.DataFrame({'요일':['월','화','수','목','금','토','일'],
                    '공부시간(시간)':[2,3,1,5,4,2,1],
                    '성적':[70,75,65,85,80,70,60]
                    })
df2
```

```python
# 2. 생성한 데이터프레임을 이용하여 요일별 공부 시간을 선 그래프로 나타내세요. x축은 요일, y축은 공부시간으로 설정하세요.

plt.plot(df2['요일'],df2['공부시간(시간)'],marker='o')
plt.rc('font', family='NanumGothic')
plt.show()
```

```python
# 3. 생성한 데이터프레임을 이용하여 성적을 선 그래프로 나타내세요. x축은 요일, y축은 성적으로 설정하세요.

plt.plot(df2['요일'],df2['성적'],'--',marker='^')
plt.rc('font', family='NanumGothic')
plt.show()
```

```python
# 생성한 데이터프레임을 이용하여 공부 시간과 성적 사이를 산점도로 나타내세요. x축은 공부시간, y축은 성적으로 설정하세요.

plt.scatter(df2['공부시간(시간)'],[df2['성적']])
plt.show()
```

```python
data_1=pd.DataFrame({
    'x':[1,2,3,4,5],
    'y':[10,20,15,25,30],
    'category':['A','B','A','B','A']
})

data_1
```

```python
sns.scatterplot(x='x',y='y',hue='category',data=data_1)
```

```python
sns.countplot(x='species',data=iris_data)
plt.show()
```

```python
# 주어진 데이터프레임에서 "category" 열은 데이터 포인트의 범주를, "value"열은 해당 데이터 포인트의 값을 나타냅니다. 주어진 데이터를 이용하여 seaborn의 countplot() 함수를 사용하여 다음을 수행하세요:

# 1. "category" 열을 사용하여 각 카테고리별로 데이터의 개수를 세어서 막대 그래프로 시각화하세요.

data_2 = pd.DataFrame({
    'category':['A','B','A','C','B','A','A','C','B','B'],
    'value':[10,20,15,25,30,12,18,22,28,35]
})

data_2
```

```python
sns.countplot(x='category',hue='category',data=data_2)
plt.show()
```
