```python
# 데이터를 이용하여 데이터프레임을 생성하세요.

df7 = pd.DataFrame({'학생':['John', 'Alice', 'Bob', 'Emily'],
                   '수학':[90,87,78,92],
                   '영어':[85,91,80,88],
                   '과학':[88,89,85,82]})
df7
```

```python
# 생성한 데이터프레임의 요약 통계 정보를 출력하세요.

df_describe = df7.describe()
df_describe
```

```python
# 생성한 데이터프레임에서 '수학' 점수가 가장 높은 학생의 정보를 출력하세요.

df7[df7['수학'] == df7['수학'].max()]
```

```python
# 생성한 데이터프레임에서 '수학' 점수가 90점 이상인 학생의 정보를 출력하세요.

math_90 = df7[df7['수학'] >= 90]
math_90
```

```python
# 역대 박스오피스 파일을 데이터프레임으로 생성하기

# 1. 데이터 파일을 생성한 후 데이터프레임의 정보를 출력하시오.

df8 = pd.read_excel('BoxOffice.xlsx')
df8
```

```python
# 대표국적별 영화 편수를 집계하시오

nation = df8['대표국적'].value_counts()
nation
```

```python
# 대표국적이 '영국'인 데이터를 출력하시오

df8[df8['대표국적']=='영국']
```

```python
# 대표국적별 최상위 순위를 출력하시오

grouped_min = df8.groupby('대표국적')['순위'].min()
print(grouped_min)
```
